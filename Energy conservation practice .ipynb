```python
#second session pylab 1

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# CONSTANTS:

delta_t = 0.01 # timestep in seconds
k = 3.78 # spring constant, in N/m
m = 0.2 # mass of the bob, in kg
amp = 7.16 # amplitude of the oscillations, in cm
t_naught = 0 # start time
T = 10.0 # end time

# initial conditions:

y_naught = amp # initial displacement value
v_naught = 0 # initial velocity value

omega_naught = np.sqrt(k/m) # constant

time_values = np.arange(t_naught, T + delta_t, delta_t) # initalizing 
# time_values array

results = pd.DataFrame(columns = ['Time', 'Position', 'Velocity', 'Energy'])
# initalizing results dataframe

y = y_naught
v = v_naught
E = 0

#  main for loop which performs the numerical integration
for i, time in enumerate(time_values):
    
    y += delta_t * v # y = y + ...
    results.loc[time, 'Position'] = y
    v -= delta_t * (omega_naught ** 2) * y 
    results.loc[time, 'Velocity'] = v
    E = (1/2) * m * (v ** 2) + (1/2) * k * (y ** 2)
    results.loc[time, 'Energy'] = E
    
results.Time = time_values # adds time to the results df

print(results)

# Plotting:

plt.figure(1, figsize = (10, 10))
plt.plot(results.Time, results.Position)
plt.xlabel('Time(s)'), plt.ylabel('Displacement(m)') # Displacement vs. Time
plt.figure(2, figsize = (10, 10))
plt.plot(results.Time, results.Velocity)
plt.xlabel('Time(s)'), plt.ylabel('Velocity(m/s)') # Velocity vs. Time
plt.figure(3, figsize = (10, 10))
plt.plot(results.Position, results.Velocity)
plt.xlabel('Displacement(m)'), plt.ylabel('Velocity(m/s)') # Velocity vs. 
# Displacement
plt.figure(4)
plt.plot(results.Time, results.Energy)
plt.xlabel('Time(s)'), plt.ylabel('Energy(J)') # Energy vs. Time
plt.show()


#2 session pylab1
```
